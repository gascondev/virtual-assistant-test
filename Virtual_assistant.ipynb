```python
# MODELOS PRE-ENTRENADOS

MODELO_VOZ_A_TEXTO = "openai/whisper-small"
MODELO_TRADUCCION = "facebook/nllb-200-distilled-600M"
MODELO_TEXTO_A_VOZ = "facebook/mms-tts-spa"
```

```python
from transformers import pipeline
import IPython.display as ipd

# Cargamos los pipelines de Hugging Face
asr_pipeline = pipeline("automatic-speech-recognition", model=MODELO_VOZ_A_TEXTO)
translator_pipeline = pipeline("translation", model=MODELO_TRADUCCION)
tts_pipeline = pipeline("text-to-speech", model=MODELO_TEXTO_A_VOZ)

# Creamos un diccionario de c√≥digos de idioma para el modelo de traducci√≥n
IDIOMAS_NLLB = {
    "ca": "cat_Latn",  # Catal√°n/valenciano
    "de": "deu_Latn",  # Alem√°n
    "en": "eng_Latn",  # Ingl√©s
    "it": "ita_Latn",  # Italiano
    "es": "spa_Latn",  # Espa√±ol
}
```

```python
# DE VOZ A TEXTO

def voz_a_texto(archivo_audio):

    resultado = asr_pipeline(archivo_audio)
    texto_transcrito = resultado["text"]
    resultado_final = print(f"üìù Texto transcrito: {texto_transcrito}")
    return resultado_final

voz_a_texto("italiano.wav")
```

```python
# TRADUCCI√ìN

def traducir_texto(texto, idioma_origen, idioma_destino):

    src_lang = IDIOMAS_NLLB.get(idioma_origen, "eng_Latn")
    tgt_lang = IDIOMAS_NLLB.get(idioma_destino, "spa_Latn")

    print(f"üîÅ Traduciendo de {src_lang} a {tgt_lang}...")

    resultado = translator_pipeline(texto, src_lang=src_lang, tgt_lang=tgt_lang)
    texto_traducido = resultado[0]["translation_text"]

    resultado_final = print(f"üìñ Traducci√≥n: {texto_traducido}")
    return resultado_final


traducir_texto("Ich arbeite gerne mit k√ºnstlicher Intelligenz", "de", "es")
```

```python
# DE TEXTO A VOZ

def texto_a_voz(texto, idioma="es"):

    print(f"üéôÔ∏è Sintetizando voz en {idioma}...")
    audio_generado = tts_pipeline(texto)

    return ipd.Audio(audio_generado["audio"], rate=audio_generado["sampling_rate"])

texto_a_voz("Hablo un poco de alem√°n.", "es")
```

```python
# ASISTENTE VIRTUAL

def asistente_virtual(archivo_audio, idioma_origen, idioma_destino):

    # 1. Transcribimos el audio
    texto = voz_a_texto(archivo_audio)

    # 2. Traducimos el texto al idioma destino
    texto_traducido = traducir_texto(texto, idioma_origen, idioma_destino)

    # 3. Convertimos el texto traducido a voz
    return texto_a_voz(texto_traducido, idioma_destino)


asistente_virtual("italiano2.wav", "it", "es") # De italiano a espa√±ol
```
